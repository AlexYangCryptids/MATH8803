```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_phi(t_list, I):
    phi = 0
    phi_list = []
    
    q = 1
    dt = t_list[1] - t_list[0]
    
    for t in t_list:
        dphi_dt = q+ I + (I-q) * np.cos(phi)
        phi += dt * dphi_dt
        phi_list.append(phi)
    return phi_list
```

```python
I = 60
T_span = 50
t_list = np.linspace(0, T_span, 1000)
dt = t_list[1] - t_list[0] 

phi_list = compute_phi(t_list, I)

print(np.where(np.array(phi_list)>2 * np.pi)[0][0] *dt)
plt.plot(t_list, phi_list)
plt.show()
```

```python
I_list = [0.05, 0.1, 0.3, 0.7, 1, 3, 5, 10, 20, 40, 60]

T_span = 100
t_list = np.linspace(0, T_span, 5000)
dt = t_list[1] - t_list[0] 

T_list = []
f_list = []

for I in I_list:
    phi_list = compute_phi(t_list, I)
    if np.where(np.array(phi_list)<2 * np.pi)[0][-1] == 4999:
        T = 10000
        f = 0
    else:
        T = np.where(np.array(phi_list)>2 * np.pi)[0][0] *dt
        f = 1/T
    
    T_list.append(T)
    f_list.append(f)

plt.plot(I_list, T_list)
plt.show()
plt.scatter(I_list, f_list, c='red')
plt.plot(I_list, f_list)
plt.xlabel('I-I_C')
plt.ylabel('freq')
plt.grid(True)
plt.show()

    
```
